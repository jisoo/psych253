```python
def plot_func_and_slopes(F, dF, xvals, title, start=-5, stop=5, start_delta=-1, stop_delta=1):

    x = np.arange(start, stop, .01)
    y = F(x)
    
    plt.plot(x, y)

    xvals = np.array(xvals)
    slopes = dF(xvals)
    intercepts = F(xvals) - dF(xvals) * xvals
    
    slope_starts_x = xvals + start_delta
    slope_ends_x = xvals + stop_delta
    
    slope_starts_y = slopes * slope_starts_x + intercepts
    slope_ends_y = slopes * slope_ends_x + intercepts
    
    things = zip(slope_starts_x, slope_ends_x, slope_starts_y, slope_ends_y)
    for (a, b, c, d) in things:
        plt.plot([a, b], [c, d]);
 
    plt.xlabel('value of x')
    plt.ylabel('value of F(x)')
    plt.title(title)
    plt.ylim(y.min()-.1, y.max() + .1)
    
```

```python
F = lambda x: np.sin(x)
dF = lambda x: np.cos(x)

plot_func_and_slopes(F, dF, 
                     xvals=[-np.pi, -np.pi/2, 0, 6*np.pi/6, 7*np.pi/4],
                     title='F(x) = sin(x)',
                     start=-2*np.pi, stop=2*np.pi)
```

## Taylor Series

```python
#the vandermonde matrix function will be useful 
np.vander([-.5, 0, .5], 3, increasing=True)
```

```python
def taylor_series(funcs, x0):
    """
    funcs = derivatives of the function to some order
    x0 = point around which the approximation is to be built
    
    returns callable computing approximation around x0
    """
    #n = order of the approximation
    n = len(funcs)
    
    #derivatives to order n at x0
    derivs = np.array([f(x0) for f in funcs])
    
    #weights of each term
    weights = 1. / sp.special.factorial(np.arange(n))

    #actually form the approximation         
    approx = lambda deltas: np.dot(np.vander(deltas, n, increasing=True),
                                   weights * derivs)

    return approx
```

```python
def taylor_series_plot(funcs, x0, start_delta=-1, stop_delta=1, eval_delta=0.01):
    
    true_func = funcs[0]
    approx_func = taylor_series(funcs, x0)
    
    xvals = np.arange(x0 + start_delta, x0 + stop_delta, eval_delta)
    yvals = true_func(xvals)
    yapprox = approx_func(xvals - x0)
    
    plt.plot(xvals, yvals)
    plt.plot(xvals, yapprox)
    
    minval = min(yapprox.min(), yvals.min())
    maxval = max(yapprox.max(), yvals.max())
    plt.ylim(minval-.1, maxval + .1)
```

```python
F = lambda x: x**2
dF = lambda x: 2 * x

plt.figure(figsize = (12, 5))
plt.subplot(1, 2, 1)
taylor_series_plot([F, dF], 0)

plt.subplot(1, 2, 2)
taylor_series_plot([F, dF], 1)
```

## Multi-variable derivatives

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
#the function
F = lambda x, y: x**2 + y**2
#its gradient function (produces a 2-vector for every input)
dF = lambda x, y: np.array([2 * x, 2 * y])

x_vals = np.arange(-2, 2, .02)
y_vals = np.arange(-2, 2, .02)
xv, yv = np.meshgrid(x_vals, y_vals)
funcval = F(xv.ravel(), yv.ravel())
gradval = dF(xv.ravel(), yv.ravel())

fig = plt.figure(figsize=(14, 4))
ax = plt.subplot(1, 3, 1, projection='3d')
ax.plot_surface(xv, yv, funcval.reshape((200, 200)), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.title('function value -- F(x, y) = x^2 + y^2')
ax = plt.subplot(1, 3, 2, projection='3d')
ax.plot_surface(xv, yv, gradval[0].reshape((200, 200)), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.title('x derivative')
ax = plt.subplot(1, 3, 3, projection='3d')
ax.plot_surface(xv, yv, gradval[1].reshape((200, 200)), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.title('y derivative')

```

```python
F = lambda x, y: x**2 + y**2 * x
dF = lambda x, y: np.array([2 * x + y**2, 2 * y * x])

x_vals = np.arange(-2, 2, .02)
y_vals = np.arange(-2, 2, .02)
xv, yv = np.meshgrid(x_vals, y_vals)
funcval = F(xv.ravel(), yv.ravel())
gradval = dF(xv.ravel(), yv.ravel())

fig = plt.figure(figsize=(14, 4))
ax = plt.subplot(1, 3, 1, projection='3d')
ax.plot_surface(xv, yv, funcval.reshape((200, 200)), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.title('function value -- F(x, y) = x**2 + y**2 * x')
ax = plt.subplot(1, 3, 2, projection='3d')
ax.plot_surface(xv, yv, gradval[0].reshape((200, 200)), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.title('x derivative')
ax = plt.subplot(1, 3, 3, projection='3d')
ax.plot_surface(xv, yv, gradval[1].reshape((200, 200)), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.title('y derivative')


```

### The hessian matrix

```python
def plot_func_and_hessians(F, dF, d2F, title='', start=-2, end=2, delta=0.02):
    x_vals = np.arange(start, end, delta)
    y_vals = np.arange(start, end, delta)
    xv, yv = np.meshgrid(x_vals, y_vals)
    funcval = F(xv.ravel(), yv.ravel())
    gradval = dF(xv.ravel(), yv.ravel())
    hessval = d2F(xv.ravel(), yv.ravel())

    N = len(x_vals)
    fig = plt.figure(figsize=(16, 8))
    ax = plt.subplot(2, 4, 1, projection='3d')
    ax.plot_surface(xv, yv, funcval.reshape((N, N)), cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    plt.title('function value -- %s' % title)
    ax = plt.subplot(2, 4, 2, projection='3d')
    ax.plot_surface(xv, yv, gradval[0].reshape((N, N)), cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    plt.title('x derivative')
    ax = plt.subplot(2, 4, 3, projection='3d')
    ax.plot_surface(xv, yv, gradval[1].reshape((N, N)), cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    plt.title('y derivative')


    ax = plt.subplot(2, 4, 5, projection='3d')
    ax.plot_surface(xv, yv, hessval[0, 0].reshape((N, N)), cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    plt.title('xx derivative')
    ax = plt.subplot(2, 4, 6, projection='3d')
    ax.plot_surface(xv, yv, hessval[0, 1].reshape((N, N)), cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    plt.title('xy derivative')
    ax = plt.subplot(2, 4, 7, projection='3d')
    ax.plot_surface(xv, yv, hessval[1, 0].reshape((N, N)), cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    plt.title('yx derivative')
    ax = plt.subplot(2, 4, 8, projection='3d')
    ax.plot_surface(xv, yv, hessval[1, 1].reshape((N, N)), cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    plt.title('yy derivative')
```

```python
F = lambda x, y: x**2 + y**2
dF = lambda x, y: np.array([2 * x, 2 * y])
d2F = lambda x, y: np.array([[2 * np.ones_like(y), np.zeros_like(y)], 
                             [np.zeros_like(y), 2 * np.ones_like(y)]]) 

plot_func_and_hessians(F, dF, d2F, title='F(x, y) = x**2 + y**2')
```

```python
#see, (0, 0) is a critical point
dF(0, 0)
```

```python
#what does the matrix of 2nd derivatives look like
#at this critical point?
d2F(0, 0)
```

```python
#determinant is positive, but diagonal is positive, so this is a local maximum
np.linalg.det(d2F(0, 0))
```

```python
F = lambda x, y: - (x**2 + y**2)
dF = lambda x, y: np.array([-2 * x, -2 * y])
d2F = lambda x, y: np.array([[-2 * np.ones_like(y), np.zeros_like(y)], 
                             [np.zeros_like(y), -2 * np.ones_like(y)]]) 

plot_func_and_hessians(F, dF, d2F, title='F(x, y) = -(x**2 + y**2)')
```

```python
#again, (0, 0) is a critical point
dF(0, 0)
```

```python
#hessians:
d2F(0, 0)
```

```python
F = lambda x, y: x**2 + y**2 * x
dF = lambda x, y: np.array([2 * x + y**2, 2 * y * x])
d2F = lambda x, y: np.array([[2 * np.ones_like(y), 2 * y], 
                             [2 * y, 2 * x]])

plot_func_and_hessians(F, dF, d2F, title='F(x, y) = x**2 + y**2 * x')
```

```python
#again, (0, 0) is a critical point
dF(0, 0)
```

```python
#hessians
d2F(0, 0)
```

```python
#determinant is 0, so can't tell -- happens to be a saddle
np.linalg.det(d2F(0, 0))
```

```python
F = lambda x, y: x**2 + y**2 * (x - 1)
dF = lambda x, y: np.array([2 * x + y**2, 2 * y * (x - 1)])
d2F = lambda x, y: np.array([[2 * np.ones_like(y), 2 * y], 
                             [2 * y, 2 * (x - 1)]])

plot_func_and_hessians(F, dF, d2F, title='F(x, y) = x**2 + y**2 * (x-1)')
```

```python
#yep it's a critical point
dF(0, 0)
```

```python
#hessians
d2F(0, 0)
```

```python
#negative determinant -- so it's definitely a saddle point
np.linalg.det(d2F(0, 0))
```
